## Initial Setup

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import time

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import Perceptron
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

df = pd.read_csv('mushroom.csv')
```

## Data Cleaning and Preprocessing

Getting the column names of categorical and numeric feature names:

```python
# Remember to one-hot encode these via pd.get_dummies()
categorical_feature_names = df.select_dtypes(include='object').columns.tolist()
# Excludes one-hot encoded categorical features
numeric_feature_names = df.select_dtypes(include='number').columns.tolist()
categorical_feature_names.remove('class')
```

Before we do any exploratory data analysis, let's deal with the missing data values:

```python
# Encode missing values in categorical features 
# with a token '?' that means "unknown".
df[categorical_feature_names] = df[categorical_feature_names].fillna('?')
# Interpolate any missing values in the numeric features.
df[numeric_feature_names] = df[numeric_feature_names].interpolate(axis=0)
```

## Exploratory Data Analysis

Let's start by analyzing the numeric features of this dataset to see if there are any outliers.

```python
numeric_instance_features = {}
for feature in numeric_feature_names:
    numeric_instance_features[feature] = df[feature]

fig, axes = plt.subplots(nrows=1, ncols=3)
for i, feature in enumerate(numeric_feature_names):
    #print(i, feature)
    sns.boxplot(data=numeric_instance_features[feature], ax=axes[i])

axes[0].set_ylabel("Cap Diameter (cm)")
axes[1].set_ylabel("Stem Height (cm)")
axes[2].set_ylabel("Stem Width (mm)")
plt.tight_layout()
plt.show()
```

## Handling Outliers

Now that we know outliers exist, let's see how the machine learning models will perform if we exclude these outliers from their input data.

We start by calculating the interquartile range (IQR) of each numeric feature, then use the IQR to calculate the upper and lower outlier bounds of each numeric feature:

```python
# Make a separate copy of the dataframe, so that
# we can compare the machine learning models'
# performance with and without outliers.
df_prune = pd.read_csv('mushroom.csv')
# Encode missing values in categorical features 
# with a token '?' that means "unknown".
df_prune[categorical_feature_names] = df_prune[categorical_feature_names].fillna('?')
# Interpolate any missing values in the numeric features.
df_prune[numeric_feature_names] = df_prune[numeric_feature_names].interpolate(axis=0)
for i, feature in enumerate(numeric_feature_names):
    data_values = np.sort( numeric_instance_features[feature].values )
    q1 = np.percentile(data_values, 25)
    q3 = np.percentile(data_values, 75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    outliers_mask = (df_prune[feature] < lower_bound) | (df_prune[feature] > upper_bound)
    df_prune = df_prune[~outliers_mask]

df_prune
```

Now that numeric outliers have been addressed, let's view how the numerical features are distributed using histograms. Start by only selecting instances which have been classified as "poisonous":

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8,10))
for i, feature in enumerate(numeric_feature_names):
    series = df_prune[df_prune['class'] == 'p'][feature]
    sns.histplot(data=series.values, ax=axes[i], kde=True)
    axes[i].set_ylabel("Poisonous Mushroom Count")

axes[0].set_xlabel("Cap Diameter (cm)")
axes[1].set_xlabel("Stem Height (cm)")
axes[2].set_xlabel("Stem Width (mm)")
plt.tight_layout()
plt.savefig('Numeric Feature Histograms.png', dpi=300)
plt.show()
```

## Data Encoding

Before feeding the data into some machine learning models, let's perform sone data encoding on the original dataframe and the dataframe after pruning outliers (since we're interested in comparing the performance of the models with and without outliers).

Data encoding on the dataframe with outliers:

```python
# Instead of one-hot encoding the predicted class (poisinous or not poisonous),
# we'll convert the letters "p" and "e" into numbers 1 and 0 so that 
# we can use them in a binary classifier algorithm.
# "p" = 1 for poison mushroom, "e" = 0 for edible mushroom (not poisonous).
df['class'] = df['class'].replace({'p':1, 'e':0})
# Perform one-hot encoding on each categorical feature.
df = pd.get_dummies(df, columns=categorical_feature_names, drop_first=False).astype(float)
```

Data encoding on the dataframe without outliers:

```python
# Numeric encoding for class labels
df_prune['class'] = df_prune['class'].replace({'p':1, 'e':0})
# One-hot encoding for categorical features
df_prune = pd.get_dummies(df_prune, columns=categorical_feature_names, drop_first=False).astype(float)
```

Split the dataset into feature data (X) and target data (y) for each dataframe:

```python
# Now split the dataset into feature data (X) and target data (y).
X = df.drop('class', axis=1)
y = df['class'].astype(int)

X_prune = df_prune.drop('class', axis=1)
y_prune = df_prune['class'].astype(int)

test_proportion = 0.2
random_seed = int(time.time())

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=test_proportion, 
                                                    random_state=random_seed, 
                                                    stratify=y)

X_train_prune, X_test_prune, y_train_prune, y_test_prune = train_test_split(X_prune, 
                                                                            y_prune, 
                                                                            test_size=test_proportion, 
                                                                            random_state=random_seed, 
                                                                            stratify=y_prune)
```

Let's perform scaling on the training and test feature sets:

```python
# Perform scaling on the training and test feature sets
# for numerical features and one-hot encoded categorical features.
column_names = X_train.columns
# I want to preserve the column names
sc = StandardScaler().set_output(transform="pandas")
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
# For cross-validation score when using Support Vector Classifier
X_std = sc.transform(X)
```

Scaling for the data sets after pruning outliers:

```python
sc = sc = StandardScaler().set_output(transform="pandas")
sc.fit(X_train_prune)
X_train_prune_std = sc.transform(X_train_prune)
X_test_prune_std = sc.transform(X_test_prune)
# For cross-validation score when using Support Vector Classifier
X_prune_std = sc.transform(X_prune)
```

## Model Training and Testing

### Linear Perceptron Classifier

Using a linear perceptron classifier on the dataset (with outliers):

```python
def test_model(X,y,model_obj):
    prediction = model_obj.predict(X)
    incorrect = (y != prediction).sum()
    correct = y.size - incorrect
    score = accuracy_score(y, prediction)
    return prediction, incorrect, correct, score

def print_results(y, incorrect, correct, score):
    print('# of samples:', y.size)
    print('# of incorrect classifications:', incorrect)
    print('# of correct classifications:', correct)
    print('Accuracy score of classifier:', score)

lpc = Perceptron(random_state=random_seed, eta0=0.1)
lpc.fit(X_train_std, y_train)
y_pred_test, incorrect_test, correct_test, y_test_score = test_model(X_test_std, y_test, lpc)
y_pred_train, incorrect_train, correct_train, y_train_score = test_model(X_train_std, y_train, lpc)
print('DATASET WITH OUTLIERS:')
print('-'*55)
print('Using the test dataset:')
print_results(y_test, incorrect_test, correct_test, y_test_score)
print('-'*55)
print('Using the training dataset:')
print_results(y_train, incorrect_train, correct_train, y_train_score)
```

Cross-validation on the linear perceptron classifier (using dataset with outliers):

```python
cross_validation_scores = cross_val_score(lpc, X_std, y, cv=5)
print(cross_validation_scores)
print("Linear Perceptron Classifier Performance:")
print("Mean cross-validation score:", cross_validation_scores.mean())
print("Standard deviation of cross-validation scores:", cross_validation_scores.std())
```

Using a linear perceptron classifier (after pruning outliers):

```python
lpc_prune = Perceptron(random_state=random_seed, eta0=0.1)
lpc_prune.fit(X_train_prune_std, y_train_prune)
y_pred_test_prune, incorrect_test_prune, correct_test_prune, y_test_score_prune = test_model(X_test_prune_std, y_test_prune, lpc_prune)
y_pred_train_prune, incorrect_train_prune, correct_train_prune, y_train_score_prune = test_model(X_train_prune_std, y_train_prune, lpc_prune)
print('DATASET WITHOUT OUTLIERS:')
print('-'*55)
print('Using the test dataset:')
print_results(y_test_prune, incorrect_test_prune, correct_test_prune, y_test_score_prune)
print('-'*55)
print('Using the training dataset:')
print_results(y_train_prune, incorrect_train_prune, correct_train_prune, y_train_score_prune)
```

Cross-validation on the linear perceptron classifier (using dataset after pruning outliers):

```python
cross_validation_scores = cross_val_score(lpc_prune, X_prune_std, y_prune, cv=5)
print(cross_validation_scores)
print("Linear Perceptron Classifier Performance:")
print("Mean cross-validation score:", cross_validation_scores.mean())
print("Standard deviation of cross-validation scores:", cross_validation_scores.std())
```

### Weight Comparison

Now let's compare the weights that the perceptron model assigns to each feature by absolute value (using the dataset with outliers intact):

```python
# Assign each feature's name to its corresponding weight:
lpc_feature_weights = {}
for i, feature in enumerate(lpc.feature_names_in_):
    lpc_feature_weights[feature] = lpc.coef_[0][i]

lpc_feature_weights_abs = {}
for feature, weight in lpc_feature_weights.items():
    lpc_feature_weights_abs[feature] = abs(weight)

# Sort the feature weights by absolute value
lpc_sorted_abs_weights = sorted(lpc_feature_weights_abs.items(), key=lambda item: item[1], reverse=True)

# Plot the 10 features with the largest absolute value

fig, ax = plt.subplots(figsize=(20,8))
for i in range(0,10):
    ax.bar(lpc_sorted_abs_weights[i][0], lpc_sorted_abs_weights[i][1])
plt.grid(axis='y')
ax.set_title('10 Features with Largest Absolute Value Weights (Using Linear Perceptron Classifier on Dataset With Outliers)')
plt.show()
```
